```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv('cleanData/usuarios.csv', sep=";")

# Filtrar las columnas que contienen información de género
gender_cols = [col for col in df.columns if 'id' not in col.lower()]

# Rellenar valores nulos con 0 y calcular la matriz de correlación
corr_matrix = df[gender_cols].fillna(0).corr()

# Visualizar la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación - Columnas de Género')
plt.show()
```
